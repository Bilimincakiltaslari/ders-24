```python
# Sadece Gonca
from turtle import * 
from random import randint
speed(5)
color("orange","green")
begin_fill()
for a in range(0,360,90):
    seth(a)
    circle(90,90)
    goto(0,0)
    circle(-90,90)
    goto(0,0)
end_fill()
color("orange","orange")
begin_fill()
penup()
goto(-10,0)
pendown()
circle(10)
end_fill()
for k in range(20):
    x = randint(-6,6)
    y = randint(-6,6)
    penup()
    goto(x,y)
    seth(0)
    pendown()
    color("white","white")
    begin_fill()
    circle(0.5)
    end_fill()
hideturtle()
mainloop()
done()
```

```python
# yaprak gonca çiçek
from turtle import * 
from random import randint
speed(5)
color("orange","green")
begin_fill()
for a in range(0,360,30):
    seth(a)
    circle(90,90)
    goto(0,0)
    circle(-90,90)
    goto(0,0)
end_fill()
penup()
goto(-10,4)
pendown()
color("red")
width(2)
begin_fill()
for a in range(15):
    fd(25)
    lt(168)
    end_fill()
penup()
goto(-1,-4)
pendown()
color("white","orange")
width(1)
begin_fill()
circle(4)
end_fill()
hideturtle()
mainloop()
done()
```

```python
# iç içe 6 renkli çember 4 lü döngü
from turtle import * 
from random import randint
speed(0)
renk=["red","green","black","indigo","brown","yellow"]
index=0
for j in range(0,360,90):
    index=int(j/90)
    if index >= 6:
        index = index - 6
    seth(j)
    for i in range(130,4,-25): 
        color("blue",renk[index])
        if index == 5:
            index = -1
        begin_fill()
        if i <= 5:
            i = 15
        circle(i)
        end_fill()
        index += 1
for k in range(50):
    x = randint(-210,210)
    y = randint(-140,140)
    yrcp = randint(3,6)
    penup()
    goto(x,y)
    seth(0)
    pendown()
    color("white","white")
    begin_fill()
    circle(yrcp)
    end_fill()
hideturtle()
mainloop()
done()
```
